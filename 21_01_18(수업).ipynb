```python
feature_name=pd.read_csv('../data/UCI HAR Dataset/human_features.txt',sep='\s+',header=None,names=['column_index','column_name'])
```

```python
feature_name=feature_name.iloc[:, 1].values.tolist()
```

```python
X_test=pd.read_csv('../data/UCI HAR Dataset/X_test.txt',sep='\s+',header=None)
X_train=pd.read_csv('../data/UCI HAR Dataset/X_train.txt',sep='\s+',header=None)
y_test=pd.read_csv('../data/UCI HAR Dataset/y_test.txt',sep='\s+',header=None)
y_train=pd.read_csv('../data/UCI HAR Dataset/y_train.txt',sep='\s+',header=None)
```

```python
y_test
```

```python
X_test.columns=feature_name
X_train.columns=feature_name
y_test=y_test.rename(columns={0:'action'})
y_train=y_train.rename(columns={0:'action'})
```

```python
y_train.isnull().sum()
```

```python
y_train.value_counts()
```

```python
X_train.isnull().sum()
```

```python
X_train=X_train.fillna(6)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
rfc = RandomForestClassifier(n_estimators=40,max_depth=5,random_state=42)
rfc.fit(X_train,y_train)
```

```python
print(rfc.score(X_train,y_train))
print(rfc.score(X_test,y_test))
```

```python
predict= rfc.predict(X_test)
predict
```

```python
print(classification_report(y_test, predict))
```

```python
from sklearn.model_selection import GridSearchCV
# random forest의 성능 향상을 위해서 옵션 설정

params={
    'n_estimators':[100],
    'max_depth':[6,8,10,12],
    'min_samples_leaf':[8,12,18],
    'min_samples_split':[8,16,20]
}
```

```python
rfc_par = RandomForestClassifier(n_jobs=-1)
```

```python
grid_rfc=GridSearchCV(rfc_par,param_grid=params,n_jobs=-1)
```

```python
grid_rfc
```

```python
grid_rfc.fit(X_train,y_train)
```

```python
print('최적 하이퍼 파라미터: \n',grid_rfc.best_params_)
print('best accuracy: {0:.3f}',format(grid_rfc.best_score_))
```
